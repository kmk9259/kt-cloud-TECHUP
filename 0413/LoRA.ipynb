# PEFT(Parameter Efficient Fine-tuning)

딥러닝 모델의 크기가 증가함에 따라, 기존과 같이 태스크에 따른 모델 튜닝에 있어 기존과 같이 모든 파라미터를 미세조정하는 full-fine-tuning 방식은 자원 측면에서 점점 불가능해지고 있다.

이에 파라미터 효율적으로 fine-tuning 하는 PEFT 방법이 활발하게 연구되고 있다.

- PEFT는 다운스트림 작업의 성능을 유지하거나 심지어 향상시키면서 학습에 필요한 계산량, 혹은 모델의 크기를 줄이는 것을 목표로 한다.



PEFT는 아래와 같은 장점을 가진다:


✔ 필요한 계산량의 감소: 모델 크기를 줄이고, 중복된 파라미터를 제거하며 정밀도를 낮춤으로써 연산량을 줄인다.

✔ 추론 속도의 향상: 모델 크기가 작아지고 계산 복잡도가 줄어듬에 따라 추론 속도가 감소하여 실시간 애플리케이션에 적합해진다.

✔ 리소스가 제한된 디바이스에서의 배포 개선: 모바일 디바이스나 엣지 디바이스와 같이 제한적인 시소스에의 배포가 용이해진다.

✔  비용 절감: fine-tuning 및 추론에 드는 모든 비용이 절감된다

✔ 선응 유지 혹은 향상: 모델 크기가 줄어들면서도 다운스트림 태스크에의 성능을 유지하거나 향상하는 것이 가능하다.

### 대표적인 PEFT 방법론 6가지

1. 지식증류(Knowledge distillation)
- 지식 증류(knowledge distrystal)는 지식을 크고 성과가 좋은 모델(teacher)에서 더 작은 모델(student)로 전달하는 기술이다. Teacher 모델의 output probability를 student 모델 학습을 위한 소프트 타깃(soft target)으로 사용한다.

- 이 과정은 student 모델이 teacher 모델이 학습한 방대한 지식을 얻을 뿐만 아니라, 일반화를 더 잘할 수 있도록 한다. 모델 크기를 줄일 뿐만 아니라 더 작은 student 모델이 정규화(regularize)하도록 도움을 주는 것이다.  



2. 모델 가지치기(Pruning)
- 가지치기는 사전 학습된 모델에서 불필요한 가중치나 연결을 제거하는 기법이다. 중복되거나 덜 중요한 매개변수를 식별하고 제거함으로써 모델의 크기와 계산 요구 사항을 크게 줄일 수 있다. Magnitude-based pruning(크기가 작은 가중치를 제거함), Structured pruning(중요도에 따라 특정 뉴런 또는 필터를 제거) 등 다양한 방법이 존재하며, 성능 저하 없이 중복 매개변수를 제거하여 효율적으로 fine-tuning 할 수 있다.



3. 양자화 (Quantization)
- 양자화는 모델 매개변수(가중치)의 정밀도를 낮춰 메모리 및 계산 요구 사항을 낮추는 기술이다다. 기존 딥러닝 모델에서 매개변수는 일반적으로 32비트 부동 소수점 숫자로 저장된다. 하지만 양자화를 사용하면 이러한 매개변수를 8비트 정수와 같이 더 낮은 비트 정밀도로 표현할 수 있다. 이렇게 정밀도를 낮추면 메모리 사용량이 크게 줄어들고 계산 속도가 빨라진다.

- 양자화는 학습 후 사전 학습된 모델을 양자화하는 post-training quantization 기법과 양자화를 염두에 두고 모델을 학습하는 quantization-aware training으로 나뉜다. 두 가지 접근 방식 모두 모델 성능을 유지하면서 효율적인 fine-tuning을 수행한다.



4. Low Rank Factorization
- Low-rank factorization은 사전학습된 모델의 가중치 행렬을 row-rank 행렬로 근사화하는 기술이다. 가중치 행렬을 저순위 구성 요소로 분해함으로써 모델의 매개 변수 수와 계산 복잡도를 줄인다. 단수 값 분해(SVD) 및 텐서 인수분해와 같은 low-rank factoraization 방법은 모델이 수행해야 하는 태스크에서 가장 중요한 정보를 유지하는 low-rank 근사치를 찾는 것을 목표로 한다. 이 기술은 합리적인 성능을 달성하면서 매개 변수 효율적인 미세 조정 옵션을 제공한다.



5. 지식 주입(Knowledge Injection)
- 지식 주입은 원래 모델의 파라미터를 수정하지 않고, 태스크 별 정보를 주입함으로써 특정 태스크에 대한 사전 학습된 모델의 성능을 향상시키는 기술이다. 이는 태스크별 모듈을 추가하거나 기존 아키텍처를 수정하여 태스크별 feature을 학습할 수 있도록 구현한다. 지식 주입은 사전 학습된 모델에 이미 존재하는 지식을 활용하면서도 재학습의 필요성을 최소화하기 때문에 효율적인 fine-tuning이 가능하다. 원본 모델에 대한 수정을 최소화하고자 하는 시나리오에서 특히 유용한 방법론이다.



6. Adapter Modules(어댑터 모듈)
- 어댑터 모듈은 원래의 파라미터를 수정하지 않고 특정 작업을 위해 사전 학습된 모델에 추가하는 경량 모듈을 말한다. 이 모듈은 태스크별로 관리해야 하며, 사전 학습된 모델의 아키텍처의 각기 다른 계층에 삽입하는 것이 가능하다. 어댑터 모듈은 사전 학습된 파라미터의 대부분을 그대로 유지하면서 모델이 작업별 정보를 학습할 수 있도록 함으로써 효율적인 미세 조정이 가능하도록 한다. 이 접근 방식은 전체 모델을 재학습하지 않고도 어댑터를 다른 태스크들에 대해 쉽게 추가하거나 제거할 수 있으므로 유연성과 효율성을 제공한다.

# LoRA란?

LoRA는 **Low-Rank Adaptation**의 줄임말로, 대규모 언어모델(LLM)을 **전체 다 다시 학습하지 않고**,  
일부 작은 파라미터만 추가로 학습해서 미세조정(fine-tuning)하는 방법.

기존 모델 전체를 뜯어고치는 대신**작은 보정 장치만 붙여서 학습하는 방식**.

---

# 왜 LoRA가 필요한가?

대형 모델은 파라미터 수가 매우 많음.  
예를 들어 수억 개, 수십억 개 파라미터를 가진 모델을 그대로 fine-tuning 하려면:

- GPU 메모리가 많이 필요함
- 학습 시간이 오래 걸림
- 저장 공간이 큼
- 실습 환경에서는 부담이 큼

LoRA는 이 문제를 줄이기 위해 나왔습니다.



- **전체 모델은 그대로 두고**
- **아주 작은 부분만 학습해서**
- **적은 자원으로도 fine-tuning 가능하게 하는 방법**

---

# 핵심 아이디어

원래 신경망의 가중치 행렬을 $ W $라고 하겠습니다.

보통 full fine-tuning은 이 $ W $ 자체를 직접 업데이트합니다.

하지만 LoRA는 $ W $를 그대로 두고, 작은 두 행렬 $ A $, $ B $를 추가해서 다음처럼 바꿉니다.

$$
W' = W + \Delta W
$$

여기서

$$
\Delta W = BA
$$


$$
W' = W + BA
$$

가 됩니다.

---

# 수식의 의미

- $ W $: 원래 모델의 가중치
- $ A $, $ B $: 새로 추가하는 작은 학습 파라미터
- $ BA $: 원래 가중치에 더해지는 보정값

핵심은 $ BA $가 **저랭크(low-rank)** 구조라는 점입니다.

큰 행렬 전체를 직접 학습하는 대신 더 작은 차원으로 압축된 변화량만 학습합니다.

그래서 학습해야 할 파라미터 수가 크게 줄어듬.

---

# 왜 "Low-Rank"인가?

행렬의 랭크(rank)는 그 행렬이 담고 있는 정보의 독립적인 차원 수를 뜻합니다.

LoRA에서는 변화량 $ \Delta W $를  
작은 랭크 $ r $를 가진 두 행렬의 곱으로 표현.

예를 들어 원래 $ W $가 $ d \times k $ 크기라면,  
LoRA는 보통 다음처럼 둡니다.

$$
A \in \mathbb{R}^{r \times k}, \quad B \in \mathbb{R}^{d \times r}
$$

그러면

$$
BA \in \mathbb{R}^{d \times k}
$$

가 되어 원래 $ W $와 같은 크기의 보정값을 만들 수 있습니다.

여기서 $ r $은 보통 $ d $나 $ k $보다 훨씬 작습니다.


큰 변화를 직접 저장하지 않고 **작은 차원의 압축된 형태로 저장**.

---

# Full Fine-Tuning과 LoRA 비교

## 1. Full Fine-Tuning

모델의 모든 파라미터를 업데이트합니다.

### 장점
- 충분한 자원이 있으면 높은 성능을 낼 수 있음
- 모델 전체를 작업에 맞게 바꿀 수 있음

### 단점
- 메모리 사용량이 큼
- 학습 시간이 오래 걸림
- 저장해야 할 파인튜닝 모델 크기도 큼

---

## 2. LoRA Fine-Tuning

원래 모델 가중치는 고정하고, 추가한 작은 행렬만 학습합니다.

### 장점
- 메모리를 적게 사용
- 학습 속도가 빠름
- 여러 태스크별 adapter를 따로 저장 가능
- 실습, 연구 초기 단계, 제한된 GPU 환경에 적합

### 단점
- 경우에 따라 full fine-tuning보다 표현력이 제한될 수 있음
- 어떤 층에 적용할지에 따라 성능 차이가 날 수 있음

---

# LoRA가 특히 좋은 이유

LoRA가 많이 쓰이는 이유는 다음과 같음.

## 1. 가볍다
큰 모델 전체를 학습하지 않으므로 GPU 메모리를 적게 씁니다.

## 2. 빠르다
학습해야 하는 파라미터 수가 적어서 업데이트가 빠릅니다.

## 3. 저장이 효율적이다
모델 전체를 새로 저장하지 않고, LoRA 파라미터만 저장하면 됩니다.

## 4. 여러 작업에 쉽게 대응 가능하다
하나의 base model 위에 태스크별로 다른 LoRA adapter를 붙여 사용할 수 있습니다.

예를 들어:

- 발표문 스타일 LoRA
- 요약용 LoRA
- 질의응답용 LoRA
- 번역용 LoRA

처럼 분리할 수 있음.

---

원래 가중치는 고정하고 작은 랭크 $r$를 가진 보조 행렬만 학습하는 구조.

````python
import torch
from datasets import Dataset
from transformers import (
    AutoTokenizer,         #모델에 맞는 토크나이저(텍스트를 숫자로 변환)를 자동으로 선택
    AutoModelForCausalLM,  # GPT/Llama 같은 텍스트 생성용 모델을 자동으로 불러옴
    Trainer,               # 복잡한 학습 루프(루프, 검증, 저장 등)를 추상화 하여 실행
    TrainingArguments,     # 학습률, 배치 크기, 저장 경로 등 하이퍼파라미터를 설정함.
    DataCollatorForLanguageModeling, # 데이터를 모델 입력에 맞게 배치(Batch) 단위로 묶어주는 역할
)
from peft import LoraConfig, get_peft_model, TaskType
#LoraConfig: 저랭크 어텝터 학습을 위한 설정값(rank, alpha 등)을 정의
#get_peft_model: 일반 모델에 LoRA 레이러를 추가하여 효율적 학습이 가능한 모델로 변환
# TaskType: 모델의 작업 유형(ex.Causal LM, seq2seq)을 지정하여 최적의 설정을 도움
````

````python
device = torch.device(
    "cuda" if torch.cuda.is_available()
    else "mps" if torch.backends.mps.is_available()
    else "cpu"
)
print("device:", device)
````

````python
model_name = "distilgpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
    model.config.pad_token_id = tokenizer.eos_token_id

model = model.to(device)
print("loaded:", model_name)
````

````python
# 학습 전 생성 함수

def generate_text(model, tokenizer, prompt, max_new_tokens=80):
    model.eval()
    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id,
        )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
````

````python
# 학습 전 결과 확인

test_prompt = """Question: What is LoRA?
Answer:"""

print(generate_text(model, tokenizer, test_prompt))

# 우리가 원하는 형식으로 예쁘게 답하지 못함 -> 여기에 LoRA를 붙임.
````

````python
lora_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    inference_mode=False,
    r=8,                         # 저랭크 차원. 작을수록 가볍지만 표현력은 줄 수 있음
    lora_alpha=16,               # 업데이트 크기 스케일
    lora_dropout=0.05,           # LoRA 경로에 들어가는 드롭아웃
    target_modules="all-linear", # 어느 선형층에 LoRA를 삽입할지 / 선형/Conv1D 모듈 전체를 대상으로 잡을 수 있음
)
````

보통 처음에는 $r$=4, 8, 16 정도를 많이 시험.

````python
# chat template 적용
def make_chat_text(example):
    messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": example["instruction"]},
        {"role": "assistant", "content": example["response"]},
    ]
    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=False
    )
    return {"text": text}

dataset = dataset.map(make_chat_text)
print(dataset[0]["text"][:500])
````

````python
# 토크나이저 함수

max_length = 256

def tokenize_function(example):
    tokenized = tokenizer(
        example["text"],
        truncation=True,
        max_length=max_length,
        padding="max_length",
    )
    tokenized["labels"] = tokenized["input_ids"].copy()
    return tokenized

tokenized_train = train_dataset.map(tokenize_function, remove_columns=train_dataset.column_names)
tokenized_eval = eval_dataset.map(tokenize_function, remove_columns=eval_dataset.column_names)

tokenized_train
````

````python
# 학습 전/후 함수
def generate_answer(model, question, max_new_tokens=80):
    model.eval()

    messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": question},
    ]

    prompt = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    inputs = tokenizer(prompt, return_tensors="pt")
    inputs = {k: v.to(model.device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id
        )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
````

````python
#LoRA 모델 생성 함수
def build_lora_model(model_id, r, lora_alpha, lora_dropout):
    model = AutoModelForCausalLM.from_pretrained(
        model_id,
        torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32
    )

    peft_config = LoraConfig(
        r=r,
        lora_alpha=lora_alpha,
        lora_dropout=lora_dropout,
        bias="none",
        task_type="CAUSAL_LM",
        target_modules="all-linear"
    )

    model = get_peft_model(model, peft_config)
    return model
````

````python
#Optuna Objetive 함수
def objective(trial):
    r = trial.suggest_categorical("r", [4, 8, 16, 32])
    lora_alpha = trial.suggest_categorical("lora_alpha", [8, 16, 32, 64])
    lora_dropout = trial.suggest_float("lora_dropout", 0.0, 0.2)
    learning_rate = trial.suggest_float("learning_rate", 1e-5, 5e-4, log=True)
    num_train_epochs = trial.suggest_int("num_train_epochs", 2, 4)

    model = build_lora_model(
        model_id=model_id,
        r=r,
        lora_alpha=lora_alpha,
        lora_dropout=lora_dropout
    )

    output_dir = f"./optuna_trial_{trial.number}"

    training_args = TrainingArguments(
        output_dir=output_dir,
        per_device_train_batch_size=1,
        per_device_eval_batch_size=1,
        gradient_accumulation_steps=4,
        learning_rate=learning_rate,
        num_train_epochs=num_train_epochs,
        eval_strategy="epoch",
        save_strategy="no",
        logging_strategy="epoch",
        report_to="none",
        fp16=torch.cuda.is_available(),
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=tokenized_train,
        eval_dataset=tokenized_eval,
        data_collator=data_collator,
    )

    trainer.train()
    eval_result = trainer.evaluate()

    eval_loss = eval_result["eval_loss"]
    trial.set_user_attr("eval_loss", float(eval_loss))

    del model
    torch.cuda.empty_cache()

    return eval_loss
````

````python
# Optuna study
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=3)
````

# QLoRA (Quantized Low-Rank Adaptation)

사전학습된 대형 언어 모델을 **4-bit 양자화한 상태로 보관**하고, **기본 모델 가중치는 고정(frozen)** 한 채, **작은 LoRA adapter만 학습**하는 미세조정(fine-tuning) 방법.


- **큰 본체 모델**은 매우 가볍게 압축해서 유지하고
- **실제로 학습되는 부분**은 작은 저랭크 보정항만 두는 방식.

---

## 핵심

**메모리 사용량을 크게 줄이면서도, full fine-tuning에 가까운 성능을 얻는 것**이 QLoRA의 핵심.

원 논문에서는 이 방법을 통해 **65B 규모 모델을 단일 48GB GPU에서 미세조정**할 수 있음을 보였다.

---

## 왜 QLoRA가 필요한가?

일반적인 full fine-tuning은 모델의 모든 가중치를 업데이트하므로 다음과 같은 비용이 매우 크다.

- 모델 파라미터 저장 메모리
- gradient 저장 메모리
- optimizer state 저장 메모리

모델 크기가 커질수록 이 세 가지가 모두 급격히 증가.

예를 들어 파라미터 수가 매우 큰 LLM에서는 단순히 모델 가중치만 올리는 것이 아니라,

- forward를 위한 메모리
- backward를 위한 intermediate activation
    - 신경망의 중간층에서 계산되어 잠시 저장되는 출력값
- optimizer가 유지하는 추가 상태값

까지 모두 필요하므로, 실제 학습 비용은 훨씬 더 커짐.

---

## LoRA와 QLoRA의 차이

### LoRA
LoRA는 **학습하는 파라미터 수를 줄이는 방법**.

기존 가중치 전체를 업데이트하지 않고, 작은 저랭크 행렬 $A, B$만 학습하여 보정항을 추가.

$$
W' = W + \Delta W = W + BA
$$

여기서

- $W$: 기존 pretrained weight
- $A, B$: 학습되는 작은 저랭크 행렬
- $\Delta W = BA$: 전체 weight update를 근사하는 보정항

---

### QLoRA
QLoRA는 LoRA의 장점에 더해, **기본 모델 자체를 4-bit로 양자화하여 저장 메모리까지 줄이는 방법**.

- **LoRA**: 학습 파라미터 수를 줄임
- **QLoRA**: 학습 파라미터 수를 줄이면서, 기본 모델 저장도 더 가볍게 함

이라고 볼 수 있다.

---

## 기본 아이디어

QLoRA의 흐름은 다음과 같다.

1. 사전학습 모델 가중치를 **4-bit로 양자화**
2. 양자화된 기본 모델은 **frozen**
3. 역전파는 통과시키되, **업데이트는 LoRA adapter만 수행**
4. 계산 안정성을 위해 일부 연산은 **더 높은 정밀도**로 수행

> 무거운 본체는 작게 압축해서 들고 있고,  
> 실제 학습은 작은 추가 부품만 바꾸는 방식

---

## 수식으로 보는 QLoRA

원래 선형층이 다음과 같다고 하자.

$$
y = Wx
$$

LoRA에서는 여기에 저랭크 보정항을 추가하여

$$
y = \left(W + \frac{\alpha}{r}BA\right)x
$$

로 바꾼다.

- $W \in \mathbb{R}^{d \times k}$: 원래 weight
- $A \in \mathbb{R}^{r \times k}$
- $B \in \mathbb{R}^{d \times r}$
- $r \ll \min(d,k)$: 작은 rank
- $\alpha$: scaling factor

QLoRA에서는 이 $W$를 full precision으로 저장하지 않고, **양자화된 형태**로 유지.

개념적으로는

$$
W \approx \hat{W}_{4bit}
$$

로 보고, 실제 forward에서는

$$
y = \left(\hat{W}_{4bit} + \frac{\alpha}{r}BA\right)x
$$

를 계산.

- $\hat{W}_{4bit}$는 **고정**
- $A, B$만 **학습**

된다는 것.

---

## 양자화는 어떻게 들어가는가?

딥러닝에서 양자화는 실수값을 더 적은 비트 수의 표현으로 근사하는 것.

예를 들어 어떤 실수 가중치 $w$를 정수값 $q$로 바꾸고 다시 근사값 $\hat{w}$로 복원하면,

$$
q = \operatorname{round}\left(\frac{w}{s}\right)
$$

$$
\hat{w} = s q
$$

처럼 생각할 수 있다.

QLoRA에서는 기본 모델 전체를 이런 방식으로 낮은 비트 폭으로 표현.

다만 실제 구현에서는 단순 정수화만이 아니라, **모델 성능 손실을 줄이기 위한 정교한 4-bit 양자화 방식**을 사용.

---

## QLoRA의 핵심 구성 요소

QLoRA를 이해할 때 중요한 구성 요소는 다음과 같다.

### 1. 4-bit NormalFloat (NF4)
QLoRA는 보통 **NF4**라는 4-bit 데이터 타입을 사용한다.

NF4는 일반적인 균등 양자화보다, **정규분포 형태의 weight 분포를 더 잘 반영하도록 설계된 4-bit 표현 방식**.

LLM의 가중치는 대체로 특정 분포를 가지므로, 이런 특성을 활용하면 단순 INT4보다 더 좋은 근사를 얻을 수 있다.

---

### 2. Double Quantization
양자화에서는 scale 같은 추가 정보도 저장해야 한다.

QLoRA는 이 scale 값들조차 다시 양자화하여 **추가 메모리까지 더 절약**.


- weight를 양자화하고
- 그 양자화에 필요한 보조 파라미터도 다시 압축하는 방식이다.

이 때문에 “4-bit로 줄였다”는 것이 단순히 weight 값만 줄인 것이 아니라, 전체 저장 구조를 매우 효율적으로 구성한 것에 가깝다.

---

### 3. Paged Optimizer
대형 모델 학습에서는 optimizer state와 순간적인 메모리 스파이크도 부담이 됨.  
QLoRA는 이런 문제를 줄이기 위해 **paged optimizer** 개념을 사용.

이는 메모리 사용량이 갑자기 튀는 상황을 완화하여, 제한된 GPU 메모리 안에서 더 안정적으로 학습하도록 돕는다.

---

## 왜 4-bit인데도 학습이 가능한가?

여기서 자주 생기는 오해는 다음과 같다.

> “기본 모델이 4-bit면 gradient 계산도 잘 안 되는 것 아닌가?”

핵심은 **기본 모델을 직접 업데이트하지 않는다는 점**.

QLoRA에서는

- 양자화된 base model은 forward 계산에 사용되고
- backward는 그 경로를 따라 흐르지만
- 실제 파라미터 업데이트는 LoRA adapter에만 적용된다.

학습의 자유도를 작게 제한하는 대신 메모리를 크게 절약하는 것이다.

---

## Full Fine-Tuning, LoRA, QLoRA 비교

| 방법 | 기본 모델 업데이트 | 추가 파라미터 | 기본 모델 저장 방식 | 메모리 사용 |
|------|-------------------|----------------|----------------------|-------------|
| Full Fine-Tuning | 전체 업데이트 | 없음 | 보통 FP16/BF16/FP32 | 가장 큼 |
| LoRA | 본체 고정, adapter만 업데이트 | 작음 | 보통 FP16/BF16 | 중간 |
| QLoRA | 본체 고정, adapter만 업데이트 | 작음 | 4-bit 양자화 | 가장 작음 |

---

## QLoRA의 장점

### 1. 매우 큰 메모리 절감
기본 모델을 4-bit로 저장하므로, pretrained model 자체의 메모리 부담이 크게 줄어듬.

### 2. 작은 GPU에서도 대형 모델 미세조정 가능
기존에는 여러 GPU나 매우 큰 GPU 메모리가 필요했던 설정을, 더 제한된 환경에서도 시도할 수 있다.

### 3. LoRA의 효율성 유지
학습 파라미터 수가 작아서 optimizer state와 gradient 메모리도 줄어듬.

### 4. 성능 대비 효율이 좋음
full fine-tuning보다 훨씬 가볍지만, 적절한 설정에서는 매우 경쟁력 있는 성능을 낼 수 있다.

---

## QLoRA의 한계

### 1. Full Fine-Tuning을 완전히 대체하는 것은 아님
모든 상황에서 full fine-tuning과 완전히 동일한 성능을 내는 것은 아니다.

특히 매우 큰 도메인 변화나 복잡한 적응이 필요할 때는 한계가 있을 수 있다.

### 2. 양자화 오차가 존재
기본 모델이 4-bit 근사값으로 저장되므로, 원래 full precision 가중치와는 차이가 있다.

### 3. 학습 가능한 부분이 제한됨
업데이트가 LoRA adapter에만 일어나므로, 모델 전체를 자유롭게 수정하는 것보다 표현력이 제한될 수 있다.

### 4. 구현 세부사항에 영향을 많이 받음
어떤 layer에 adapter를 붙이는지, rank를 얼마로 둘지, 학습률을 어떻게 설정하는지에 따라 결과가 꽤 달라질 수 있다.

---

## QLoRA에서 자주 조절하는 하이퍼파라미터

### 1. rank $r$
LoRA 보정항의 크기를 결정한다.

$$
\Delta W = BA
$$

에서 $A, B$의 중간 차원인 $r$이 클수록 더 풍부한 보정이 가능하지만, 학습 파라미터 수와 메모리도 증가.

---

### 2. LoRA alpha
보정항의 크기를 조절하는 스케일 파라미터.

$$
W' = W + \frac{\alpha}{r}BA
$$

에서 $\alpha$가 너무 작으면 adaptation이 약하고, 너무 크면 불안정할 수 있다.

---

### 3. target modules
Transformer의 어떤 선형층에 LoRA를 붙일지 정한다.

보통 다음과 같은 attention projection에 많이 적용.

- `q_proj`
- `k_proj`
- `v_proj`
- `o_proj`

경우에 따라 MLP layer까지 포함하기도 한다.

---

### 4. 학습률, batch size, sequence length
QLoRA도 결국 미세조정이므로, 일반적인 fine-tuning 설정의 영향을 크게 받는다.  
특히 sequence length가 길어지면 activation memory가 다시 커질 수 있다.

---

## Transformer에서 QLoRA가 붙는 위치

Transformer의 attention에서 보통

$$
Q = W_Q x, \quad K = W_K x, \quad V = W_V x
$$

처럼 projection을 계산.

QLoRA에서는 예를 들어 $W_Q$, $W_V$에 대해

$$
W_Q' = \hat{W}_{Q,4bit} + \frac{\alpha}{r}B_QA_Q
$$

$$
W_V' = \hat{W}_{V,4bit} + \frac{\alpha}{r}B_VA_V
$$

처럼 두고, base weight는 양자화된 채로 고정하고 adapter만 학습한다.

---

- **기본 모델**: 아주 크고 무거운 기계
- **양자화**: 그 기계를 가능한 한 가볍게 압축해서 보관
- **LoRA adapter**: 기계 전체를 뜯어고치지 않고, 작은 조절 장치만 추가
- **학습**: 그 작은 조절 장치만 조정
  
**큰 본체는 그대로 두고, 작은 보정 장치만 훈련해서 목적에 맞게 적응시키는 방식**.

---

## 언제 QLoRA를 쓰면 좋은가?

QLoRA는 다음과 같은 상황에서 특히 유용하다.

- GPU 메모리가 제한적일 때
- 큰 LLM을 직접 full fine-tuning하기 어려울 때
- 특정 태스크나 도메인에 빠르게 적응시키고 싶을 때
- 여러 adapter를 따로 관리하며 실험하고 싶을 때

예를 들어,

- 의료 문서 요약
- 법률 문서 질의응답
- 고객지원 챗봇
- 특정 스타일 답변 생성

````python
# 모델 지정
model_id = "Qwen/Qwen2.5-1.5B-Instruct"
````

````python
# 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained(model_id, use_fast=True)

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

print("pad_token:", tokenizer.pad_token)
print("eos_token:", tokenizer.eos_token)
````

````python
#chat template 형태로 학습용 텍스트 만들기
## instruct 모델은 대화 형식이 잘 맞음

def make_chat_text(example):
    messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": example["instruction"]},
        {"role": "assistant", "content": example["response"]},
    ]

    text = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=False
    )
    return {"text": text}

dataset = dataset.map(make_chat_text)
dataset[0]["text"][:500]
````

- bit 양자화 설정
    - load_in_4bit=True
    - bnb_4bit_quant_type="nf4"
    - bnb_4bit_use_double_quant=True
    - bnb_4bit_compute_dtype=torch.bfloat16

    
조합을 제시. 또한 NF4는 4-bit 학습용으로 권장

````python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16
)
````

````python
# 모델 로드
## 학습시에는 device_map="auto"를 주지 않음. /  추론용이기 때문에

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    trust_remote_code=True
)

model.config.use_cache = False
````

````python
# k-bit training
# 양자화된 모델을 학습 전에 prepare_model_for_kbit_training()으로 전처리하라고 안내
model = prepare_model_for_kbit_training(model)
````

````python
# LoRA 설정
# PEFT 문서 예시는 특정 projection 레이어를 지정하지만, 실습에서는 더 편하게 all-linear를 씀.
peft_config = LoraConfig(
    r=8,
    lora_alpha=16,
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
    target_modules="all-linear"
)
````

````python
# LoRA 모델 만들기
model = get_peft_model(model, peft_config)
model.print_trainable_parameters()
````

**전체 모델이 아니라 일부 adapter만 학습**

````python
# 학습전 출력 함수
def generate_answer(question, max_new_tokens=80):
    messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": question},
    ]

    prompt = tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
    )

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id
        )

    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return decoded

test_question = "QLoRA가 무엇인지 한 문장으로 설명해줘."
print(generate_answer(test_question))
````
