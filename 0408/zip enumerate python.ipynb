```python
def my_decorator(func):
    def wrapper():
        print("함수 실행 전")
        func()
        print("함수 실행 후")
    return wrapper

@my_decorator
def say_hello():
    print("안녕하세요")

say_hello()

```

```python
say_hello = my_decorator(say_hello)
```

 - class에 적용하는 데코레이터
    - @staticmethod
    - @classmethod
    - @propery


 - @staticmethod
    - 클래스 안에 있지만 self를 받지 않는 함수
    - self가 없어서 상속과 같은, 아무것도 전달이 되지 않음

```python
class Calculator:
    
    @staticmethod
    def add(a,b):
        return a+b
print(Calculator.add(3,5))
```

 - @classmethod
    - 첫 번째(self) -> cls로 받음
    

```python
class Student:
    school = "ABC university"

    @classmethod
    # 객체 자체가 아니라 클래스에 대한 정보를 받음
    # 현재 클래스를 사용하여 객체를 생성
    def show_school(cls):
        print(cls.school)

Student.show_school()
```

```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    @classmethod
    def from_string(cls,data):
        name,age = data.split(",")
        return cls(name, int(age))
p = Person.from_string("youngpyo,30")
print(p.name,p.age)
```

 - @property
    - 메소드를 변수처럼 접근하게 만들어줌

```python
class Circle:
    def __init__(self,radius):
        self.radius = radius

    @property
    def area(self):
        return 3.14 * self.radius**2
c = Circle(3)
print(c.area)
```

```python
# 상속
def add_tag(cls):
    cls.tag = "base"
    return cls

@add_tag
class Base:
    pass

class Child(Base):
    pass

print(Base.tag)
print(Child.tag)
```
