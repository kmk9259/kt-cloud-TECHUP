03/20 숙제: 사이트 하나를 정해서 크롤링을 하여 전처리 및 토큰화까지 마친 후 Corpus를 제작해보기.

```python
# 1. 특정 사이트 url에서 BeautifulSoup으로 텍스트 가져오기
import requests # http 라이브러리
from bs4 import BeautifulSoup # html 데이터 추출 라이브러리
from kiwipiepy import Kiwi # kiwi 형태소 분석 라이브러리

# 검색 페이지 URL
url = "https://ktcloud-techup.goorm.io"

res = requests.get(url, timeout=10) # get으로 웹페이지 불러오기
if res.status_code == 200: # status 200으로 정상이라면
    soup = BeautifulSoup(res.content, "html.parser") # html parser로 response content 파싱

    # 2. 과정별 추천 강의 가져오기
    # list comprehension: [결과값 for 변수 in 조건문]
    allCoursesList = [  
        title.get_text(strip=True) # course title text 추출
        for title in soup.find_all("div", class_="_14ksIk card-title") # div중에서 _14ksIk card-title class 모두 찾기
    ]
    print(allCoursesList)

    #3. 토큰화 & 형태소 분석
    kiwi = Kiwi() # kiwi 생성자
    kiwi.add_user_word('NLP', 'NNG')
    stopwords = {'직장인'} # 불용어 정의, set
    
    processed = [
        [token.form for token in kiwi.tokenize(title) #2. 토큰화한 내용을 token.form이라고 할 때
         if token.tag in {'NNG', 'NNP'} and token.form not in stopwords]   #3. token.tag가 명사이면서 token.form이 불용어가 아닌 token.form을 담아라
        for title in allCoursesList #1. allCoursesList의 title에서
    ]
    for i, nouns in enumerate(processed, 1): # iterable 객체를 순회하면서 idx와 element를 모두 가져옴
        print(f"{i}번째 문장: {nouns}")
else:
    print("fail")
```
