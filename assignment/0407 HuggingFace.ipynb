1. HuggingFace Hub 살펴보기
    - 모델이나 목적을 정해서 간략하게 어떤 것인지 서술
        - BERT라는 모델을 더 가볍고 빠르게 만든 DistilBERT를 기반으로 하는 모델
        - SST-2라는 데이터를 학습하여 문장이 긍정, 부정 판별하는 파인튜닝됨
    - 이 모델이 어떤 작업에 쓰이는지?
        - 텍스트 분류 중에서도 문장 감정 분석 작업에 사용
    - 모델 페이지에서 중요한 키워드?
        - transformer
        - fine-tuned
2. 모델 불러오기
    - transformers 설치
    - pipeline 이나 autotokenizer, automodel 아무거나 써보기
    - input를 몇 개 넣어서 결과 확인
3. 결과 정리
4. 느낀 점
    - hugging face의 장단점
        - 장점: 접근성과 표준화가 잘되어 있다.
        - 단점: 아직까지는 영어 중심인 모델이 많다? 잘 모르겠다

```python
from transformers import pipeline

# 1. 모델 불러오기 (pipeline 사용)
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

# 2. Input 데이터 넣기
inputs = [
    "I am so happy to learn how to use Hugging Face!",
    "This homework is a bit challenging, but I can do it.",
    "I don't like waiting for the model to download."
]

# 3. 결과 확인
results = classifier(inputs)

for text, result in zip(inputs, results):
    print(f"Text: {text}")
    print(f"Result: {result['label']}, Score: {result['score']:.4f}\n")
```
